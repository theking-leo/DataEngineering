Now let's loop through json["list"] to get the temperature etc.... for each 3 hour time period

```
three_hour_forecast_dict = {"city": [],
                            "temperature": [],
                            "description": [],
                            "wind_speed": [],
                            "rain": []}

for item in json["list"]:
  three_hour_forecast_dict["city"].append(json["city"]["name"])
  three_hour_forecast_dict["temperature"].append(item['main']["temp"])
  three_hour_forecast_dict["description"].append(item["weather"][0]["description"])
  three_hour_forecast_dict["wind_speed"].append(item["wind"]["speed"])
  if "rain" in item.keys():
    three_hour_forecast_dict["rain"].append(item["rain"]["3h"])
  else:
    three_hour_forecast_dict["rain"].append(0)

pd.DataFrame(three_hour_forecast_dict)
```

```
cities = ["Berlin", "Hamburg"]
three_hour_forecast_dict = {"city": [],
                            "temperature": [],
                            "description": [],
                            "wind_speed": [],
                            "rain": []}

for city in cities:

  url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_key}"

  response = requests.get(url)
  json = response.json()

  for item in json["list"]:
    three_hour_forecast_dict["city"].append(json["city"]["name"])
    three_hour_forecast_dict["temperature"].append(item['main']["temp"])
    three_hour_forecast_dict["description"].append(item["weather"][0]["description"])
    three_hour_forecast_dict["wind_speed"].append(item["wind"]["speed"])
    if "rain" in item.keys():
      three_hour_forecast_dict["rain"].append(item["rain"]["3h"])
    else:
      three_hour_forecast_dict["rain"].append(0)

pd.DataFrame(three_hour_forecast_dict)
```

Time to make a function

```
cities = ["Berlin", "Hamburg"]
```

```
def weather_by_city(cities):

  three_hour_forecast_dict = {"city": [],
                              "temperature": [],
                              "description": [],
                              "wind_speed": [],
                              "rain": []}
  for city in cities:

    url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_key}"
    response = requests.get(url)
    json = response.json()

    for item in json["list"]:
      three_hour_forecast_dict["city"].append(json["city"]["name"])
      three_hour_forecast_dict["temperature"].append(item['main']["temp"])
      three_hour_forecast_dict["description"].append(item["weather"][0]["description"])
      three_hour_forecast_dict["wind_speed"].append(item["wind"]["speed"])
      if "rain" in item.keys():
        three_hour_forecast_dict["rain"].append(item["rain"]["3h"])
      else:
        three_hour_forecast_dict["rain"].append(0)

  return pd.DataFrame(three_hour_forecast_dict)
```

```
weather_by_city(cities)
```
